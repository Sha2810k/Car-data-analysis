# SHOW WHICH TRANSMISSION TYPE HAS HIGHEST RATNGS ?

```python
grouped = df.groupby('transmission_type')
```

```python
mean_ratings = grouped['rating'].mean()
```

```python
highest_rated_transmission = mean_ratings.idxmax()
```

```python
print("The highest rated transmission type is:", highest_rated_transmission)
```

# HOW AVERAGE PRICE VARIES WITH BODY TYPE

---



```python
df['Average Price'] = (df['starting_price'] + df['ending_price']) / 2
```

```python
average_price_per_body_type = df.groupby('body_type')['Average Price'].mean()
```

```python
average_price_per_body_type.plot(kind='bar')
```
